```python
import numpy as np
ratings = np.array([10,9,8,7,6,5,4,3])
np.sort(np.power(ratings / 10, 2))[::-1]
```

```python
import matplotlib.pyplot as plt

plt.plot(ratings, np.power(ratings / 10, 1))   # not create nonlinear emphasis on high ratings
plt.plot(ratings, np.power(ratings / 10, 1.5)) # lowest rating is still have large weight (~0.2)
plt.plot(ratings, np.power(ratings / 10, 2))   # -> suitable
plt.plot(ratings, np.power(ratings / 10, 3))   # a bit extreme
plt.title('Rating powers')
plt.show()
```
